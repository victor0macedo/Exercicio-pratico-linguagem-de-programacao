## 7. Análise por canal

### Pergunta de negócio
**Qual canal de venda é o mais eficiente?**

### Tarefa
Crie uma tabela agregada por `canal_venda` contendo, no mínimo:
- Receita Total
- Lucro Total
- Margem Média ou Margem Calculada
- Quantidade de pedidos
- Ticket Médio

### Depois responda
1. Qual canal gera mais receita?
2. Qual canal gera maior eficiência?
3. Se você fosse gestor, priorizaria volume ou margem?


```python
analise_canal = df.groupby('canal_venda').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('pedido_id', 'nunique')
).reset_index()

analise_canal['margem_calculada'] = (analise_canal['lucro_total'] / analise_canal['receita_total']) * 100
analise_canal['ticket_medio'] = analise_canal['receita_total'] / analise_canal['quantidade_pedidos']

analise_canal['receita_total'] = analise_canal['receita_total'].map('R$ {:,.2f}'.format)
analise_canal['lucro_total'] = analise_canal['lucro_total'].map('R$ {:,.2f}'.format)
analise_canal['margem_calculada'] = analise_canal['margem_calculada'].map('{:,.2f}%'.format)
analise_canal['ticket_medio'] = analise_canal['ticket_medio'].map('R$ {:,.2f}'.format)

display(analise_canal.sort_values(by='receita_total', ascending=False))
```

## 10. O motor da análise: agregações (groupby)

Nos slides, a lógica central é:
- `receita -> sum`
- `margem -> mean` ou cálculo derivado
- `produto -> count`
fileciteturn5file0

### Tarefa
Escolha **duas dimensões diferentes** e construa tabelas agregadas mostrando que você entendeu essa lógica.

```python
print("\n--- Análise por UF ---")
analise_uf = df.groupby('uf').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('pedido_id', 'nunique')
).reset_index()

analise_uf['margem_calculada'] = (analise_uf['lucro_total'] / analise_uf['receita_total']) * 100
analise_uf['ticket_medio'] = analise_uf['receita_total'] / analise_uf['quantidade_pedidos']

analise_uf['receita_total'] = analise_uf['receita_total'].map('R$ {:,.2f}'.format)
analise_uf['lucro_total'] = analise_uf['lucro_total'].map('R$ {:,.2f}'.format)
analise_uf['margem_calculada'] = analise_uf['margem_calculada'].map('{:,.2f}%'.format)
analise_uf['ticket_medio'] = analise_uf['ticket_medio'].map('R$ {:,.2f}'.format)

display(analise_uf.sort_values(by='receita_total', ascending=False))

print("\n--- Análise por Produto ---")
analise_produto = df.groupby('produto').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum')
).reset_index()

analise_produto['margem_calculada'] = (analise_produto['lucro_total'] / analise_produto['receita_total']) * 100

analise_produto['receita_total'] = analise_produto['receita_total'].map('R$ {:,.2f}'.format)
analise_produto['lucro_total'] = analise_produto['lucro_total'].map('R$ {:,.2f}'.format)
analise_produto['margem_calculada'] = analise_produto['margem_calculada'].map('{:,.2f}%'.format)

display(analise_produto.sort_values(by='receita_total', ascending=False).head(10)) # Mostrando apenas os Top 10 para conciseness
```
